```python
def classify_position(pos):
    if pd.isna(pos):
        return "Unknown"
    
    main_pos = pos.split(',')[0].strip()
    
    if main_pos in ['CB', 'RB', 'LB', 'LWB', 'RWB', 'LCB', 'RCB']:
        return 'Defender'
    elif main_pos in ['CM', 'CDM', 'CAM', 'LM', 'RM', 'LAM', 'RAM', 'LCM', 'RCM', 'LDM', 'RDM']:
        return 'Midfielder'
    elif main_pos == 'GK':
        return 'Goalkeeper'
    else:
        return 'Forward'

fifa['position_class'] = fifa['player_positions'].apply(classify_position)
```

### 3.3 Outliers

I am going to do it using IQR (the range between the first quartile (25th percentile) and the third quartile (75th percentile) of the data)

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"{column}: Found {len(outliers)} outliers")
    return outliers
```

```python

columns_to_check = ['log_value_eur', 'wage_eur', 'log_release_clause_eur', 'age']

outliers_dict = {}
for col in columns_to_check:
    outliers_dict[col] = detect_outliers_iqr(fifa, col)
```
